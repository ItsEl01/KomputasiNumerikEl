**Matplotlib Inline** berguna untuk memastikkan plot bisa ditampilkan langsung.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import numpy.polynomial.polynomial as poly
```

dalam kode diatas berfungsi sebagai pembuat 50 titik random dari persamaan x dan y yang sebelumnya telah ditentukan lalu diimplementasikan membentuk sebuah grafik

```python
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)
plt.scatter(x, y);
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)

model.fit(x[:, np.newaxis],y)

xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit);
```

```python
print ("Model Slope :" , model.coef_[0])
print ("Model Intercept :" , model.intercept_)
```

Pada kode diatas nilai x dan y menghasilkan nilai random dengan pola non linear dengan fungsi sinus, lalu mengeksekusi curve fitting dengan polinomial derajat 6 guna mendekati pola data tersebut.

Koefisian polinomial dihitung menggunakanpoly.polyfit dan kurva polynomial di deskripsikan dengan poly>Polynomial dan np.linspace untuk menciptakan titik baru dan y_new divisualisasikan ke dalam grafik

```python
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = np.sin(x) + 0.1 * rng.randn(50)
degree = 6
coefs = poly.polyfit(x, y, degree)
ffit = poly.Polynomial(coefs)
x_new = np.linspace(0, 10, 100)
y_new = ffit(x_new)

plt.scatter(x, y, label='Data Asli')
plt.plot(x_new, y_new, color = 'black', label = 'Kurva Polinomial')
plt.legend()
plt.title('CurveFitting dengan Polinomial')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
