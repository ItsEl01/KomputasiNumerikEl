```python
import numpy as np
import matplotlib.pyplot as plt
```

**import numpy as np** adalah library phyton


**import matplotlib.pyplot as plt**
matplotplib digunakan untuk fungsi visualisasi grafik data

```python
def f(x,y):
    return -2 * x * y
x0, y0 = 0, 1
h = 0.1
x_end = 2

x_values = np.arange(x0, x_end + h, h)
y_values = np.zeros(len(x_values))
y_values[0] = y0

for i in range(1, len(x_values)):
    x_prev = x_values[i - 1]
    y_prev = y_values[i - 1]
    y_values[i] = y_prev + h * f(x_prev, y_prev)

plt.plot(x_values, y_values, label='Euler (numerik)', marker = 'o')
```

Menyelesaikan persamaan diferensial **dy/dx = -2xy** menggunakan metode Euler.

Pertama, fungsi **f(x, y)** didefinisikan untuk merepresentasikan persamaan diferensial.

Kemudian loop **for** digunakan untuk menghitung nilai-nilai y secara iteratif berdasarkan rumus metode Euler : **y_i + 1 = y_i + h * f(x_i, y_i)**, dengan h sebagai ukuran langkah

NIlai-nilai x dan y yang dihitung disimpan dalam array **x_values** dan **y_values**, yang kemudian di plot untuk memvisualisasikan solusi numerik.

Pada dasarnya, kode ini mensimulasikan solusi persamaan diferensial dengan memperkirakan perubahan kecil pada y untuk setiap langkah kecil dalam x

```python
def analytical_solution(x):
    return np.exp(-x**2)

y_exact = analytical_solution(x_values)
plt.plot(x_values, y_exact, label='Solusi Analitik', linestyle='--')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Metode Euler untuk dy/dx = -2xy')
plt.legend()
plt.grid()
plt.show()
```

fungsi **analytical_solution(x)** didefinisikan untuk menghitung solusi analytic, yaitu exp (-x^2).

Kemudian, solusi analitik dihitung untuk setiap nilai x dalam **x_values** dan disimpan dalam **y_exact**.

Selanjutnya, solusi analitik diplot bersama dengan solusi numerik yang diperoleh sebelumnya.

Label, judul, legenda, dan grid ditambahkan di plot untuk meningkatkan kejelasan.

Akhirnya, **plt.show()** digunakan untuk menampilkan plot.
